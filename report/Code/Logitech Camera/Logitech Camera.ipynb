```python
import matplotlib.pyplot as plt
import numpy as np

def generate_camera_frame():
    # Create a simple gradient to simulate a camera feed (720p resolution)
    frame = np.zeros((720, 1280, 3), dtype=np.uint8)
    
    # Simulating a simple gradient or light effect that might be seen in a camera feed
    for i in range(720):
        frame[i, :, :] = [i % 256, 128, 255]  # Generate a gradient effect across the image
    
    return frame

def show_image(image):
    # Display the image in a window
    plt.imshow(image)
    plt.axis('off')  # Hide axes for clarity
    plt.show()

def main():
    # Generate a "camera-like" frame
    camera_frame = generate_camera_frame()
    
    # Display the simulated camera image
    show_image(camera_frame)

if __name__ == "__main__":
    main()
```
